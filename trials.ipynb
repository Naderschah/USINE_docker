```python
import os, re
import numpy as np

files = ['class_TUDatime','propag_minimisation','usine_e_options','usine_i_options']
additives = ['.ref', '.fail']

basepath = os.path.abspath('..')
```

```python
for i in files:
    print(i)
    ref = open(os.path.join(basepath,i+additives[0])).readlines()
    print('opened')
    with open(os.path.join(basepath,i+additives[1])) as second:
        fail = second.readlines()
    print('comparing')
    for j in range(len(ref)):
        if ref[j] != fail[j]:
            print(i, j, ref[j], fail[j])
            comparison_array = np.array(re.findall(r"[-+]?\d*\.\d+|\d+", ref[j]),dtype=float)-np.array(re.findall(r"[-+]?\d*\.\d+|\d+", fail[j]),dtype=float)
            if [i for i in comparison_array  if i!= 0] != []:
                print([i for i in comparison_array  if i!= 0])
                print()
    print('New File \n\n')
```
